```python
import pandas as pd
```

```python
path = "/content/weather.csv"
```

```python
df = pd.read_csv(path)
df
```

```python
df.pivot(index='date',columns='city')
```

I want only humidity to show. So i'll use the **values** parameter.

```python
df.pivot(index='date',columns='city',values='humidity')
```

using **pivot_table()**

```python
path = "/content/weather2.csv"
df= pd.read_csv(path)
df
```

In this dataframe we can see that there are two values of humidity and temperature for a day [temp and hum - morning and afternoon ]. I wanted to get the average of them. I'll use pivot table

```python
df.pivot_table(index='city',columns='date')
```

Using **aggfunc** in pivot_table()

```python
df.pivot_table(index='city',columns='date',aggfunc='sum')
```

adding **margins** to pivot_table

```python
df.pivot_table(index='city',columns='date',aggfunc='sum',margins=True)
```

Using **Grouper**()

```python
path = "/content/weather3.csv"
df=pd.read_csv(path)
df
```

The temperature and humidity of New York city is given for different months. I wanted to get the average temperature and humidity by month. So i'll use the grouper()

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```
