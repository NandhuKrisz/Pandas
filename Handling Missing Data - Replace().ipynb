#### What If there are two spatial values? -99999 and -88888

```python
df = pd.read_csv(r"C:\Users\lenovo\Desktop\Nandhu\Datasets\py-master\pandas\6_handling_missing_data_replace\weather_data.csv")  #na_values=[-99999] it can be also used
df
```

```python
df.replace([-99999,-88888],np.NaN)
```

We can see there are 0 in event column, we can replace 0 with NaN but if we do then it's gonna affect all other columns. For eg. value 0 in windspeed column is valid. So how can we do? **Using Dictionary**

```python
df.replace({
    'temperature':[-99999,-88888],
    'windspeed':[-99999,-88888],
    'event':0},np.NaN)
```

#### mapping

```python
df.replace({
    -99999:np.NaN,
    -88888:np.NaN
})
```

In this dataset the temperature contains string and also integer values. So how to deal with it? **regex (regular expression)**
